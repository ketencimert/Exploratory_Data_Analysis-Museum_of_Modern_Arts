```python
import csv
import pandas as pd

dic = {}
with open("Countries-List.csv", encoding="ISO-8859-1") as f:
    file= csv.DictReader(f, delimiter=',')
    for line in file:
        dic[line['Demonym 1']] = line['Name']
        if line['Demonym 2']:
            dic[line['Demonym 2']] = line['Name']
        if line['Demonym 3']:
            dic[line['Demonym 3']] = line['Name']
        
       
```

```python

art = pd.read_csv("MOMA_art_cleaned2.csv")
art["region1"] = art["0"].str.split().str[0].replace(dic)
art["region"] = art["region1"].str.lower()
```

```python
art["region1"] = art["region1"].replace("Russian Federation", "Russia")
art["region"] = art["region"].replace("russian federation", "russia")

art["region1"] = art["region1"].replace("United States of America", "USA")


art.head()
```
